# Convolutional Neural Network (CNN) en TensorFlow (spanglish version)

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Se carga un conjunto de datos pre-existentes de TensorFlow, en este caso es el dataset "fashion_mnist". 

Fashion-MNIST is a dataset of Zalando's article images consisting of a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes.

Los datasets de imágenes se presentan con la imagen y con un indicador que las identifica, por lo tanto son dos variables que hay que identificar al importar un dataset. Se requieren datos para realizar el aprendizaje y datos para testear qué tan bien aprendió el algoritmo, en este caso se utiliza el mismo dataset pero se identifica de forma distinta casa grupo, "train" entrena el algoritmo, "test" testea el algoritmo.

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

print(train_labels)
```

La variable "train_labels" nos identifica el tipo de prenda que corresponde a cada elemento del dataset, por lo que se debe de especificar el tipo de elementos que hay en el dataset. Para el dataset seleccionado se que se tienen los siguientes tipos de prenda

```python
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Skirt','Sneaker','Bag','Ankle boot']
```

Tamaño del dataset

```python
type(train_images)
```

Ejemplo de un elemento del dataset

```python
plt.figure()
plt.imshow(train_images[15000])
plt.grid(True)
```

Se reduce el tamaño del dataset de entrenamiento con la intención de que el dataset de testeo se lleve a cabo en un dataset más grande.

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
print(train_images.shape)
print(test_images.shape)
#print(test_images.type)
```

Ejemplo de 25 prendas del dataset con su respectiva etiqueta

```python
%matplotlib inline
plt.figure(figsize = (10,10))
for i in range(25): 
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(train_images[i],cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
```

# Modelo de predicción: Sequential model.

Un modelo secuencial es el que se usa para problemas donde a cada una de las capas del database le corresponde únicamente un tensor de input (la imagen) y un tensor de output (la etiqueta).

## Dense Neural Network (DNN)

-layers.Dense: Dense implementation is based on a large 512 unit layer followed by the final layer computing the softmax probabilities for each of the 10 categories corresponding to the 10 labels. Another implementation could be "LeNet-5 implementation".

-layers.Flatten: The input image of size 28x28 pixels is transformed into a vector in the Flatten layer, giving a feature space of width 784.

-Activation: In artificial neural networks, the activation function of a node defines the output of that node given an input or set of inputs. 
    --relu: Applies the rectified linear unit activation function.
    --softmax: Softmax converts a real vector to a vector of categorical probabilities.

```python
model = keras.Sequential([keras.layers.Flatten(input_shape = (28, 28)), keras.layers.Dense(512, activation = tf.nn.relu), keras.layers.Dense(10, activation = tf.nn.softmax)])
```

```python
model.summary()
```

Compilación del modelo. The compile() method: specifying a loss, metrics, and an optimizer

-optimizer: To improve the accuracy and reduce the loss, we need to train the neural networks by using optimization algorithms. The optimizers are one of the main components of model training. Neural network optimization is a process to fit the model with training data by adjusting the weights to get the best performance.  
    --Adam: Modelo de optimización. Otros de ellos son Adadelta, Adagrad, Adamax, Ftrl, ...
    
-loss: In the context of an optimization algorithm, the function used to evaluate a candidate solution (i.e. a set of weights) is referred to as the objective function.

We may seek to maximize or minimize the objective function, meaning that we are searching for a candidate solution that has the highest or lowest score respectively.

Typically, with neural networks, we seek to minimize the error. As such, the objective function is often referred to as a cost function or a loss function and the value calculated by the loss function is referred to as simply “loss”. Funciones de "loss": MeanSquaredError(), KLDivergence(), CosineSimilarity(), etc.

-metrics: A metric is a function that is used to judge the performance of your model. Metricas: AUC, Precision, Recall, ...

```python
model.compile(optimizer = tf.optimizers.Adam(), 
              loss = 'sparse_categorical_crossentropy', 
              metrics = ['accuracy'])
```

Entrenamiento. epoch representa el número de repeticiones del proceso, el aprendizaje se conserva en el epoch posterior.

```python
model.fit(train_images, train_labels, epochs = 5)
```

Evaluación del modelo

```python
test_loss, test_acc = model.evaluate( test_images, test_labels )
print('Accuracy ', test_acc)
```

Predicción del modelo

```python
predictions = model.predict(test_images)
```

Se plotea la predicción realizada por el modelo a partir de las test_images. Se muestra la predicción en azul si fue correcta y en rojo si fue incorrecta, entre paréntesis la etiqueta correcta.

```python
plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    predicted_label = np.argmax(predictions[i])
    true_label = test_labels[i]
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    plt.xlabel('{} ({})'.format(class_names[predicted_label], class_names[true_label]), color = color)
    
```

## Convolutional Neural Network (CNN)

There are two convolutional layers based on 3x3 filters with average pooling. The feature space is thus reduced from 32 x 32 x 3 down to 6 x 6 x 16. They are followed by 2 hidden and dense layers of 120 and 84 neurons, and finally the same 10 neuron softmax layer to compute the probabilities. Total number of coefficients of the LeNet-5 is 101 770, a quarter of the Dense DNN.

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.fashion_mnist.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Skirt','Sneaker','Bag','Ankle boot']
```

```python
#model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
# tf.keras.layers.Conv2D(
#     filters, kernel_size, strides=(1, 1), padding='valid',
#     data_format=None, dilation_rate=(1, 1), groups=1, activation=None,
#     use_bias=True, kernel_initializer='glorot_uniform',
#     bias_initializer='zeros', kernel_regularizer=None,
#     bias_regularizer=None, activity_regularizer=None, kernel_constraint=None,
#     bias_constraint=None, **kwargs
# )

modelLeNet = keras.Sequential()
modelLeNet.add(keras.layers.Conv2D(32, (3,3), padding = "same", activation='relu', input_shape=(28, 28, 1)))
modelLeNet.add(keras.layers.MaxPooling2D((2,2)))
modelLeNet.add(keras.layers.Conv2D(64 , (3,3) , activation='relu'))
modelLeNet.add(keras.layers.MaxPooling2D((2,2)))
modelLeNet.add(keras.layers.Conv2D(64, (3,3) , activation='relu'))
```

```python
#Después de haber fitteado datos
modelLeNet.summary()
```

```python
modelLeNet.add(keras.layers.Flatten())
modelLeNet.add(keras.layers.Dense(64, activation='relu'))
modelLeNet.add(keras.layers.Dense(10))
#modelLeNet.add(keras.layers.Dense(10))
```

```python
modelLeNet.summary()
```

```python
train_images = train_images.reshape(-1, 28, 28, 1)

modelLeNet.compile(optimizer='adam',
               loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
               metrics=['accuracy'])

# modelLeNet.compile(optimizer = tf.optimizers.Adam(), 
#               loss = 'sparse_categorical_crossentropy', 
#               metrics = ['accuracy'])

history = modelLeNet.fit(train_images, train_labels, batch_size = 28, epochs = 3)
#                    validation_data=(test_images, test_labels))
```

```python
test_images = test_images.reshape(-1,28,28,1)
test_loss, test_acc = modelLeNet.evaluate( test_images, test_labels )
print('Accuracy ', test_acc)
```

```python
predictions = modelLeNet.predict(test_images)
predictions.shape
```

```python
predicted_label = np.zeros(len(predictions))

for i in range(len(predictions)):
    predicted_label[i] = int(np.argmax(predictions[i]))

plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(test_images[i],cmap=plt.cm.binary)
    predicted = int(predicted_label[i])
    true_label = test_labels[i]
    if predicted == true_label:
        color = 'blue'
    else:
        color = 'red'
    plt.xlabel('{} ({})'.format(class_names[predicted], class_names[true_label]), color = color)
```

## Matriz de confusión 

```python
predicted_label = predicted_label.astype(int)

cnf_matrix = tf.math.confusion_matrix(
    test_labels, predicted_label, num_classes=None, weights=None, dtype=tf.dtypes.int32,
    name=None
)
print(cnf_matrix)
```

```python
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(cnf_matrix, annot = True, cmap = 'Blues_r', fmt='g')
ax.xaxis.set_label_position('top')
plt.tight_layout()
plt.title('Matriz de confusión', y = 1.1)
plt.ylabel('Etiqueta actual')
plt.xlabel('Etiqueta de predicción')
plt.show()
```
